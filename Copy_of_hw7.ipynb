```python
# Assignment 7   Datascience class Rajat Jhanjee

# question 0
#  Write a python function that takes an X and Y vector and returns an estimated
# b0 and b1
# and from linear regression.
# Check your function versus one of the prewritten ones.
#Include this in code cells of your ipython notebook.

import numpy as np
from scipy import stats as st
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
## this sets some style parameters
sns.set()

# generate a random sample of data of 100 elements between 0 and 1

X = np.random.uniform(size = 100)
Y = np.random.uniform(size = 100)

# calculate the mean of X and Y

x_mean = np.mean(X)
y_mean = np.mean(Y)

# calculate b0 and b1

beta1 = np.sum((X - x_mean) * (Y - y_mean)) / np.sum((X - x_mean) ** 2)
beta0 = y_mean - b1 * x_mean

print(" beta0 and beta1 from calculated linear regression ")

# print beta0 and beta1 upto three decimal places

print(f"beta0 = {beta0:.3f}")
print(f"beta1 = {beta1:.3f}")


sns.scatterplot(x=X, y=Y)
## add a line
sns.lineplot(x=X, y=beta0 + beta1 * X)

# create a regression model using stats from scipy

slope, intercept, r_value, p_value, std_err = st.linregress(X, Y)

print(" beta0 and beta1 from buitin linear regression ")

# print slope and intercept upto three decimal places

print(f"beta0 = {intercept:.3f}")
print(f"beta1 = {slope:.3f}")

sns.scatterplot(x=X, y=Y)
## add a line
sns.lineplot(x=X, y=intercept + slope * X)


```

The regression through the origin model is given by $Y = \beta X$. The ordinary least squares (OLS) estimator for $\beta$ is:

$$\hat{\beta} = \frac{\sum_{i=1}^n x_i y_i}{\sum_{i=1}^n x_i^2}$$

If the independent variable $X$ is replaced by $cX$, where $c$ is a positive constant, the new model can be written as $Y = \beta^* (cX)$. The new OLS estimator for $\beta^*$ would be:

$$\hat{\beta}^* = \frac{\sum_{i=1}^n (c x_i) y_i}{\sum_{i=1}^n (c x_i)^2}$$

Simplifying the expression:

$$\hat{\beta}^* = \frac{c \sum_{i=1}^n x_i y_i}{c^2 \sum_{i=1}^n x_i^2} = \frac{1}{c} \left( \frac{\sum_{i=1}^n x_i y_i}{\sum_{i=1}^n x_i^2} \right)$$

Therefore, the new estimate $\hat{\beta}^*$ is related to the original estimate $\hat{\beta}$ by the equation:

$$\hat{\beta}^* = \frac{\hat{\beta}}{c}$$

This shows that the estimate of the new coefficient ($\beta^*$) is the original estimate ($\beta$) divided by the constant $c$.

$Y' = \frac{Y - \bar{Y}}{\text{std}(Y)}$

$X' = \frac{X - \bar{X}}{\text{std}(X)}$


Properties of  standard vectors: mean = 0, Standard deviation = 1

Regression of Y on X is:

$b_{YX}=\frac{\text{cov}(X,Y)}{\text{var}(X)}$

Similarly, Regression of standard variables Y' on X' is:

$b_{Y^{\prime }X^{\prime }}=\frac{\text{cov}(X^{\prime },Y^{\prime })}{\text{var}(X^{\prime })}$

Since Var(X') = 1,

$b_{Y^{\prime }X^{\prime }}=\frac{r_{xy}}{1}=r_{xy}$

$b_{XY}=\frac{\text{cov}(X,Y)}{\text{var}(Y)}$
